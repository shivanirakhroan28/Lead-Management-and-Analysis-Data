```python
dummies= pd.get_dummies(df,drop_first=True,dtype="int")
```

```python
dummies
```

```python
x= dummies.drop("Converted",axis=1)
y=dummies['Converted']
```

```python
cols= x.columns
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mn= MinMaxScaler()
```

```python
x= mn.fit_transform(x)
```

```python
x= pd.DataFrame(x,columns=cols)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,random_state=32,shuffle=True,test_size=0.2,stratify=y)
```
